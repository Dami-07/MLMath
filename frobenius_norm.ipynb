```python
import numpy as np
```

The __Frobenius__ norm of a matrix __n x d__ is defined as follows

$
\|x\|_f = \|x\|^T_f = \sqrt{\sum_{i=1}^{n} \sum_{j=1}^{d} a^2_{ij}}
$

In ML is usually refered as the __energy__ of a matrix

```python
def frobenius_norm(m : np.ndarray) -> float:
    if len(m.shape) != 2:
        return 0
    else:
        mt = m.transpose()
        norm = 0
        for i in range(mt.shape[0]):
            norm += np.sum(mt[i , :]**2)
        return norm ** (1/2)
```

```python
x = np.array([[1,2],[3,4]])
print(frobenius_norm(x))
```
