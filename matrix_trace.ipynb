```python
import numpy as np
```

The trace of a square matrix is the sum of all the elements in it's diagonal and it is denotated by __tr(A)__

The trace of a __n x n__ matrix is defined as:

$\mathrm{tr}(A) =  \sum_{i=0}^{n} a_{ii}$

```python
def trace_matrix(m : np.ndarray) -> float:
    if m.shape[0] != m.shape[1] or len(m.shape) != 2:
        raise Exception("Matrix must be a square matrix")
    else:
        tr = 0
        tr += m[0,0]
        for i in range(1, m.shape[0]):
            tr += m[i , i]
        return tr
```

```python
x = np.array([[1,2],[3,4]])
print(trace_matrix(x))
```
