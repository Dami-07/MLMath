```python
import numpy as np
from math import sqrt , cos , fabs
```

The dot product or inner product is an operation defined for two vector that shares the same size
it takes 2 vector and it return the sum of the multiplication component-wise

$$
\langle \mathbf{v} , \mathbf{u} \rangle = \sum_{i=1}^n v_iu_i
$$

It can be defines with another formula , the dot product of two vector is the product of the vector module by the cosine of the angle between the two

$$
\langle \mathbf{v} , \mathbf{u} \rangle = \|v\|\|u\|cos(\theta)
$$

The module of a vector is the squared sum of it's components

$$
\|v\| = \sqrt{\sum_{i=1}^n v_i^2}
$$

The dot product has the following property:

    1)commutative
    2)distributive

```python
def module(v : np.ndarray) -> float:
    mod = sum([i ** 2 for i in v])
    return sqrt(mod)
```

```python
def dot_product(v : np.ndarray , u : np.ndarray) -> float:
    pr = sum([i*j for i,j in zip(u,v)])
    return pr
def dot_product_cosine(v : np.ndarray , u : np.ndarray , angle: float) -> float:
    return module(v)*module(v)*cos(angle)
```

A vector can also be normalized by scaling it by it module so it is normalized by unit length
the process of normalization does not change the direction of the vector but only it module,
the generalized formula is:

$$
\|v\|_p = (\sum_{i=1}^d |x_i|^p)^{(1/p)}
$$

When the parameter p is set to 1 it's called the __Manhttan norm__ or $L_1-norm$

```python
def normalize(v : np.ndarray , p : int) -> float:
    norm = sum(fabs(i)**p for i in v)**(1/p)
    return norm
```

The distance between two point in the euclidean-space is defined as follows:

$$
\|\mathbf{v}-\mathbf{u}\|^2 = \langle (\mathbf{v}-\mathbf{u}) , (\mathbf{v}-\mathbf{u}) \rangle = \sum_{i=1}^n (\mathbf{v}-\mathbf{u})^2
$$

The Cauchy-Schwarz inequality give the dot product of two vector an upper bound of the product of the module of the two vector

$$
|\sum_{i=1}^n {v_iu_i}| = \langle \mathbf{v} \mathbf{u} \rangle \le \|v\|\|u\|
$$

The cosine of the angle between two vector can be obtained through this formula:

$$
cos(\theta) = \frac{\langle \mathbf{v} , \mathbf{u} \rangle}{\|v\|\|u\|}
$$
